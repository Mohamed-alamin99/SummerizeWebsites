```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import display, Markdown
import google.generativeai as genai
```

```python
#connecting the api 
load_dotenv(override=True)
api_key = os.getenv("GOOGLEAI_API_KEY")

#validate gemini key

if not api_key :
    print("api key not fount ")
elif not api_key.startswith("AIzaS"):
    print("api key found but it seems not a gemini key")
elif api_key.strip() != api_key :
    print("api key found but it has misleading spaces")
else:
    print("Key looks good")
```

```python
# now lets call our model and print its repsponse 
genai.configure(api_key=api_key)

model = genai.GenerativeModel("gemini-2.5-flash-lite")
response = model.generate_content("tell me fun fact about egypt .")
print(response.text)
```

```python
# now it time to designn a system prompt and in this case will use chat_start
chat = model.start_chat() # initailizing the chat
system_prompt = "you are a helpful assistant that helps user to learn about egypt"
chat.send_message(system_prompt)
response = chat.send_message("tell me fun fact about egypt .")
print(response.text)

# you can use the combined function to combined system and user prompt in one line
# combined_prompt
# Send system + user prompt together
chat.send_message("""
You are a helpful assistant that speaks in a friendly tone and gives concise answers.
User: What's the capital of Egypt?
""") # ann example of combined query 
```

```python
# now it time to summerize website contents 
#already designed a function to fetch website contents
# it will called from scraper.py a function named fetch_website_contents
url = "https://web.vodafone.com.eg/en/home"
voda = fetch_website_contents(url)
print(voda)
```

```python
# Lets Build now 
def summarize(url):
    website = fetch_website_contents(url)
    model = genai.GenerativeModel("gemini-2.5-flash-lite")
    chat = model.start_chat()
    system_prompt = "you are helpful in summerization tasks specially in summerize  telecoms websites."
    chat.send_message(system_prompt)
    response = chat.send_message("Please summerize below contents ." + website )
    return response.text
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))

display_summary("https://web.vodafone.com.eg/en/home")
```
